### Using seaborn to plot the top 10 nationalities of the players

```python
#using seaborn to plot the top 10 nationalities of the players
sns.barplot(x=df['Nationality'].value_counts().head(10).index, y=df['Nationality'].value_counts().head(10).values,palette='RdGy_r')
plt.title('Top 10 Nationalities of the Players')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.show()
#using seaborn to plot the top 10 nationalities of the players
```

### Name of top 5 players with assist

```python
df[['Assists','Name']].sort_values(ascending=False,by='Assists',ignore_index=True)[:5]
```

### ploting a graph of the top 5 players with the most assists

```python
#plot a graph of the top 5 players with the most assists
df[['Assists','Name']].sort_values(ascending=False,by='Assists',ignore_index=True)[:5].plot(kind='bar',figsize=(10,5),x='Name')
plt.title('Top 5 Players with the Most Assists')
plt.xlabel('Player')
plt.ylabel('Assists')
plt.show()
#putting the names at the x axis
```

### Who scored higest penalty goals in the dataset

```python
#who scored higest penalty goals in the dataset
df[['Penalty_Goals',"Name","Club"]].sort_values(ascending=False,by='Penalty_Goals')[:1]
```

### Higest goal scorer in each club showing the name and club with goals

```python
#higest goal scorer in each club showing the name and club with goals
df[['Goals',"Name","Club"]].sort_values(ascending=False,by=['Goals']).groupby('Club').head(1)
```

### using seaborn to plot the higest goal scorer in each club showing the name and club with goals
